При одномерном сплитовании юзер попадает только в один эксперимент одновременно. Разные эксперименты проводятся последовательно для одних юзеров и могут встать в очередь. Плюс - они изолированы и просто сделать.

При многомерном эксперименте несколько эксп-тов могут проводиться одновременно. При этом для перовго эксп-та делается одно сплитование на тест и контроль, а для второго - другое из тех же юзеров, в итоге в тест и контроль из второго эксп-та имеется равные вероятности попадения юзеров из теста и кнотроля из первого эксп-та. В рез-те в каждой выборке: тест и контроль второго эксп-та содержится равное кол-во случайных юзеров из первого эксп-та и вклад 1-ого эксп-та на второй эксп-т одинаков. Поэтому можно проводить 2 эксп-та параллельно.
